```python
import re
import json
import requests

from tqdm import tqdm

from bs4 import BeautifulSoup 
from googlesearch import search 
```

```python
track_titles = []

with open('data/TechHouse_track_titles.txt', 'r', encoding='utf-8') as infile:
    track_titles = infile.read().split("\n")
```

```python
def google_search(query):
    
    flag = False
    
    for link in search(query, num=10, pause=1): 

        if re.search(r"beatport", link):
            flag = True
            #print("Beatport link found!")
            break
            
    if flag:
            
        return link
    
    else:
        print("Search failed for query: {}".format(query))

def get_soup(url):

    html = requests.get(url).content
    bsObj = BeautifulSoup(html, 'lxml')
    
    return bsObj
    
def get_track_dict(url):
    
    bsObj = get_soup(url)

    items = bsObj.find('ul', {'class': 'interior-track-content-list'}).find_all('li')

    track_dict = {}

    for item in items:

        category = item.find('span', {'class': 'category'}).string
        val = item.find('span', {'class': 'value'}).string

        track_dict[category] = val
        
    return track_dict
```

## Get the Beatport links

```python
links = []
problems = []

for query in track_titles:
    
    link = google_search(query)
    
    if link:   
        links.append(link)
        
    else:       
        problems.append(query)
        
valid_queries = [title for title in track_titles if title not in problems]
```
